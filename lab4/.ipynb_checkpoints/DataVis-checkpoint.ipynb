```r
library(tidyverse) 
library(ggplot2)
library(dplyr)
library(reshape)
library(ggpubr)
library(cowplot)
```

```r
Canada_economic_indicators <- read.csv('Canada_economic_indicators.csv')
Canada_economic_indicators <- transform(Canada_economic_indicators, Year = as.character(Canada_economic_indicators$Year))
colnames(Canada_economic_indicators) <- c("Geography", "Year", "Expenditure based GDP (2012 chained C$)", "Expenditure based GDP (current C$)", "Income based GDP (current C$)", "Industry GDP (current C$)", "Industry GDP (chained C$)", "CPI all-items", "CPI gasoline", "Labour force", 'Employment', 'Unemployment', 'Unemployment rate', 'Population', "Expenditure based GDPpC (2012 chained C$)", "Expenditure based GDPpC (current C$)", "Income based GDPpC (current C$)", "Industry GDPpC (current C$)", "Industry GDPpC (chained C$)")
Canada_economic_indicators$'Unemployment rate' <- Canada_economic_indicators$'Unemployment' / Canada_economic_indicators$"Labour force"

Canada_economic_indicators_pct <- read.csv('Canada_economic_indicators_pct.csv')
Canada_economic_indicators_pct <- transform(Canada_economic_indicators_pct, Year = as.character(Canada_economic_indicators$Year))
colnames(Canada_economic_indicators_pct) <- c("Geography", "Year", "Expenditure based GDP (2012 chained C$)", "Expenditure based GDP (current C$)", "Income based GDP (current C$)", "Industry GDP (current C$)", "Industry GDP (chained C$)", "CPI all-items", "CPI gasoline", "Labour force", 'Employment', 'Unemployment', 'Unemployment rate', 'Population', "Expenditure based GDPpC (2012 chained C$)", "Expenditure based GDPpC (current C$)", "Income based GDPpC (current C$)", "Industry GDPpC (current C$)", "Industry GDPpC (chained C$)")
Canada_economic_indicators_pct$'Unemployment rate' <- Canada_economic_indicators$'Unemployment' / Canada_economic_indicators$"Labour force"
```

```r
options(repr.plot.width=24, repr.plot.height=10)

total_gdp <- Canada_economic_indicators %>%
    filter(Year == 2019) %>%
    select(Geography, "Expenditure based GDP (current C$)") %>%
    gather(key = "variable", value = "value", -Geography) %>%
    ggplot() +
    aes(y = reorder(Geography, value), x = value) +
    geom_col(aes(fill = ifelse(Geography == "British Columbia" | Geography == "Outside Canada" | Geography == "Canada", "Highlighted", "Normal"))) + geom_text(aes(label=scales::dollar(value),fontface='bold', size= 16,hjust=0)) + 
    labs(title='2019 GDP contribution by province/territory', subtitle='BC, the 4th biggest province', caption = 'Source: Statistics Canada. Table 36-10-0222-01 Gross domestic product, expenditure-based, provincial and territorial, annual (x 1,000,000)', x = 'CA$ (MM) (current)') +
    scale_x_continuous(labels = scales::label_dollar()) + 
    theme_minimal()  + theme( 
        plot.title = element_text(size = 30, face= 'bold'),
        plot.subtitle = element_text(size = 20, face= 'bold'),
        plot.caption= element_text(size = 14),
        legend.position = 'none',
        panel.grid.major = element_blank(), 
        panel.grid.minor = element_blank(), 
        panel.background = element_blank(),
        axis.title.y=element_blank(),
        axis.title.x=element_text(size=12),
        axis.ticks.y=element_blank(),
        axis.ticks.x=element_blank(),
        axis.text.y=element_text(size=18),
        axis.text.x=element_blank()) 

total_gdp

ggsave('total_gdp.png', width = 24, height = 10)
```

```r
gdp_evo <- Canada_economic_indicators %>%
    filter(Geography == "British Columbia", Year <= 2019) %>%
    select(Year, "Expenditure based GDP (current C$)") %>%
    gather(key = "variable", value = "value", -Year) %>%
    ggplot() +
    aes(x = Year , y = value) +
    geom_area(aes(fill = variable, group=1)) + geom_text(aes(label=scales::dollar(value),fontface='bold', size= 18, vjust=-1)) + 
    labs(title='2010-2019 BC GDP evolution', subtitle='Positive trend and economic expansion', y = 'CA$ (MM) (current)', caption = 'Source: Statistics Canada. Table 36-10-0222-01 Gross domestic product, expenditure-based, provincial and territorial, annual (x 1,000,000)') +
    scale_y_continuous(labels = scales::label_dollar()) + 
    theme_minimal()  + theme( 
        plot.title = element_text(size = 30, face= 'bold'),
        plot.subtitle = element_text(size = 20, face= 'bold'),
        plot.caption= element_text(size = 14),
        legend.position = 'none',
        panel.grid.major = element_blank(), 
        panel.grid.minor = element_blank(), 
        panel.background = element_blank(),
        axis.title.x=element_blank(),
        axis.title.y=element_text(size=12),
        axis.ticks.y=element_blank(),
        axis.ticks.x=element_blank(),
        axis.text.y=element_text(size=10),
        axis.text.x=element_text(size=18)) 

gdp_evo

ggsave('gdp_evo.png', width = 24, height = 10)
```

```r
gdp_growth <- Canada_economic_indicators_pct %>%
    filter(Geography == "British Columbia", Year > 2010, Year < 2020) %>%
    select(Year, "Expenditure based GDP (current C$)") %>%
    gather(key = "variable", value = "value", -Year) %>%
    ggplot() +
    aes(x = Year , y = value) +
    geom_col(aes(fill = ifelse(value > .05, "Highlighted", "Normal"), group=1)) + geom_text(aes(label=scales::percent(value),fontface='bold', size= 18, vjust=-1)) + 
    labs(title='2011-2019 BC yearly GDP growth rates', caption = 'Source: Statistics Canada. Table 36-10-0222-01 Gross domestic product, expenditure-based, provincial and territorial, annual (x 1,000,000)', subtitle='4 years growing at +5% rate') +
    scale_y_continuous(labels = scales::label_percent()) + 
    theme_minimal()  + theme( 
        plot.title = element_text(size = 30, face= 'bold'),
        plot.subtitle = element_text(size = 20, face= 'bold'),
        plot.caption= element_text(size = 14),
        legend.position = 'none',
        panel.grid.major = element_blank(), 
        panel.grid.minor = element_blank(), 
        panel.background = element_blank(),
        axis.title.x=element_blank(),
        axis.title.y=element_blank(),
        axis.ticks.y=element_blank(),
        axis.ticks.x=element_blank(),
        axis.text.y=element_text(size=10),
        axis.text.x=element_text(size=18)) 

gdp_growth

ggsave('gdp_growth.png', width = 24, height = 10)
```
